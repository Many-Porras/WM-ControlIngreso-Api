# [dbo].[sp_ActualizarMontoSubvencion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ActualizarMontoSubvencion]    Script Date: 4/03/2025 10:31:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_ActualizarMontoSubvencion]
    @idComite INT,
    @idUsuario INT,
    @monto DECIMAL(18,2)
AS
BEGIN
    SET NOCOUNT ON;

    -- Verificar que el usuario tenga rol "administradorWM" y perfil "superAdministradorWM"
    IF NOT EXISTS (
        SELECT 1
        FROM USUARIO_WM u
        INNER JOIN ROL_WM r ON u.idRol = r.idRol
        INNER JOIN PERFIL_WM p ON r.idPerfil = p.idPerfil
        WHERE u.idUsuario = @idUsuario
          AND r.nombreRol = 'administradorWM'
          AND p.nombrePerfil = 'superAdministradorWM'
    )
    BEGIN
        PRINT 'Usuario no autorizado para actualizar el monto.';
        RETURN;
    END

    -- Verificar si ya existe un presupuesto para ese comité y usuario
    IF EXISTS (
        SELECT 1 
        FROM PRESUPUESTO_WM 
        WHERE idComite = @idComite 
          AND idUsuario = @idUsuario
    )
    BEGIN
        -- Actualizar el monto existente
        UPDATE PRESUPUESTO_WM
        SET montoPresupuesto = @monto, fechaUpdate = GETDATE()
        WHERE idComite = @idComite 
          AND idUsuario = @idUsuario;
    END
    ELSE
    BEGIN
        PRINT 'No se encontró el registro de presupuesto para este comité y usuario.';
    END
END;
GO
```

# [dbo].[sp_ListarSubvenciones]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ListarSubvenciones]    Script Date: 4/03/2025 10:31:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_ListarSubvenciones]
    @Mes INT = NULL,         -- Mes (1=Enero, 2=Febrero, etc.)
    @Anio INT = NULL         -- Año (ej. 2025)
AS
BEGIN
    SET NOCOUNT ON;

    WITH Meses AS (
    SELECT 1 AS Mes UNION ALL SELECT 2 UNION ALL SELECT 3 UNION ALL 
    SELECT 4 UNION ALL SELECT 5 UNION ALL SELECT 6 UNION ALL 
    SELECT 7 UNION ALL SELECT 8 UNION ALL SELECT 9 UNION ALL 
    SELECT 10 UNION ALL SELECT 11 UNION ALL SELECT 12
)
SELECT 
    m.Mes,
    r.nombreRegion AS Region,
    p.nombreProvincia AS Provincia,
    d.nombreDistrito AS Distrito,
    c.nombreColegio AS Colegio,
    com.idComite AS idComite,
    com.nombreComite AS Comite,
    ISNULL(pres.montoPresupuesto, 0) AS Monto,
    CASE 
        WHEN pres.montoPresupuesto IS NULL THEN 'Ingresar monto'
        ELSE 'Editar monto'
    END AS Estado
FROM Meses m
CROSS JOIN COMITE_WM com
INNER JOIN COLEGIO_WM c ON com.idColegio = c.idColegio
INNER JOIN DISTRITO_WM d ON c.idDistrito = d.idDistrito
INNER JOIN PROVINCIA_WM p ON d.idProvincia = p.idProvincia
INNER JOIN REGION_WM r ON p.idRegion = r.idRegion
LEFT JOIN PRESUPUESTO_WM pres 
    ON com.idComite = pres.idComite
    AND DATEPART(MONTH, pres.fechaHoraPresupuesto) = m.Mes
    AND (@Anio IS NULL OR DATEPART(YEAR, pres.fechaHoraPresupuesto) = @Anio)
	where m.mes = @Mes
ORDER BY m.Mes, com.nombreComite;
END;

GO
```

# [dbo].[sp_LoginUsuarioByRolPerfil]

```sql
/****** Object:  StoredProcedure [dbo].[sp_LoginUsuarioByRolPerfil]    Script Date: 4/03/2025 10:31:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_LoginUsuarioByRolPerfil]
    @numeroDocumentoUsuario VARCHAR(20),
    @passwordUsuario VARCHAR(255),
    @idPerfil INT = NULL,  -- Opcional: para restringir por perfil
    @idRol INT = NULL      -- Opcional: para restringir por rol
AS
BEGIN
    SET NOCOUNT ON;

    SELECT 
        u.idUsuario,
        u.nombreUsuario,
        u.apellidoUsuario,
        u.numeroDocumentoUsuario,
        r.idRol,
        r.nombreRol,
        p.idPerfil,
        p.nombrePerfil,
        u.fechaCreacion
    FROM USUARIO_WM u
    INNER JOIN ROL_WM r ON u.idRol = r.idRol
    INNER JOIN PERFIL_WM p ON r.idPerfil = p.idPerfil
    WHERE u.numeroDocumentoUsuario = @numeroDocumentoUsuario
      AND u.passwordUsuario = @passwordUsuario
      AND (@idPerfil IS NULL OR p.idPerfil = @idPerfil)
      AND (@idRol IS NULL OR r.idRol = @idRol);
END;
GO
```

# [dbo].[sp_RegistrarMontoSubvencion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_RegistrarMontoSubvencion]    Script Date: 4/03/2025 10:31:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_RegistrarMontoSubvencion]
    @idComite INT,
    @idUsuario INT,
    @monto DECIMAL(18,2),
    @Mes INT = NULL,      -- Opcional: Mes (1=Enero, 2=Febrero, etc.)
    @Anio INT = NULL      -- Opcional: Año (ej. 2025)
AS
BEGIN
    SET NOCOUNT ON;

    -- Verificar que el usuario tenga rol "administradorWM" y perfil "superAdministradorWM"
    IF NOT EXISTS (
        SELECT 1
        FROM USUARIO_WM u
        INNER JOIN ROL_WM r ON u.idRol = r.idRol
        INNER JOIN PERFIL_WM p ON r.idPerfil = p.idPerfil
        WHERE u.idUsuario = @idUsuario
          AND r.nombreRol = 'administradorWM'
          AND p.nombrePerfil = 'superAdministradorWM'
    )
    BEGIN
        PRINT 'Usuario no autorizado para registrar el monto.';
        RETURN;
    END

    DECLARE @fecha DATETIME;
    IF @Mes IS NOT NULL AND @Anio IS NOT NULL
        -- Se construye la fecha usando el día actual del sistema
        SET @fecha = DATEFROMPARTS(@Anio, @Mes, DAY(GETDATE()))
    ELSE
        SET @fecha = GETDATE();

    INSERT INTO PRESUPUESTO_WM (idComite, idUsuario, montoPresupuesto, fechaHoraPresupuesto, fechaCreacion)
    VALUES (@idComite, @idUsuario, @monto, @fecha, GETDATE());
END;

GO
```
